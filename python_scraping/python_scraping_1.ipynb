## 5.2 Put things together

```python
job_num = 0
url = 'https://sfbay.craigslist.org/d/software-qa-dba-etc/search/sof'
while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    jobs = soup.find_all('p', {'class': 'result-info'})
    
    # print current page
    for job in jobs:
        job_num += 1
        post_time = job.find('time', {'class': 'result-date'}).get('title')
        title = job.find('a', {'class': 'result-title'}).text
        job_link = job.find('a', {'class': 'result-title'}).get('href')
        location_tag = job.find('span', {'class': 'result-hood'})
        location_text = location_tag.text[2:-1] if location_tag else 'N/A'    

        # parse each job description page
        job_response = requests.get(job_link)
        job_data = job_response.text
        job_soup = BeautifulSoup(job_data, 'html.parser')
        ## job description
        job_description = job_soup.find('section', {'id': 'postingbody'}).text
        ## job attributes
        job_attribute_tag = job_soup.find('p', {'class':'attrgroup'})
        job_attributes = job_attribute_tag.text if job_attribute_tag else 'N/A'

        print('Job Title: ' + title + '\nPost Time: ' + post_time + '\nJob Link: ' + job_link + '\nLocation:' + location_text + '\nJob Description: ' + job_description + '\nMore Job Details: ' + job_attributes)
        
    # get url of next page
    url_tag = soup.find('a', {'title': 'next page'})
    if url_tag.get('href'):
        url = 'https://sfbay.craigslist.org' + url_tag.get('href')
    else:
        break
print('Total number of jobs: ' + str(job_num))
```

# 6. Save Scraped Data to File

```python
import pandas as pd
```

```python
npo_jobs = {}
job_num = 0
url = 'https://sfbay.craigslist.org/d/software-qa-dba-etc/search/sof'
while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    jobs = soup.find_all('p', {'class': 'result-info'})
    
    # print current page
    for job in jobs:
        job_num += 1
        post_time = job.find('time', {'class': 'result-date'}).get('title')
        title = job.find('a', {'class': 'result-title'}).text
        job_link = job.find('a', {'class': 'result-title'}).get('href')
        location_tag = job.find('span', {'class': 'result-hood'})
        location_text = location_tag.text[2:-1] if location_tag else 'N/A'    

        # parse each job description page
        job_response = requests.get(job_link)
        job_data = job_response.text
        job_soup = BeautifulSoup(job_data, 'html.parser')
        ## job description
        job_description = job_soup.find('section', {'id': 'postingbody'}).text
        ## job attributes
        job_attribute_tag = job_soup.find('p', {'class':'attrgroup'})
        job_attributes = job_attribute_tag.text if job_attribute_tag else 'N/A'

        ## update dictionary
        npo_jobs[job_num] = [title, location_text, post_time, job_attributes, job_description]
#         print('Job Title: ' + title + '\nPost Time: ' + post_time + '\nJob Link: ' + job_link + '\nLocation:' + location_text + '\nJob Description: ' + job_description + '\nMore Job Details: ' + job_attributes)
        
    # get url of next page
    url_tag = soup.find('a', {'title': 'next page'})
    if url_tag.get('href'):
        url = 'https://sfbay.craigslist.org' + url_tag.get('href')
    else:
        break
print('Total number of jobs: ' + str(job_num))
npo_jobs_df = pd.DataFrame.from_dict(npo_jobs, orient = 'index', columns = ['Job Title', 'Location', 'Post Time', 'Job Attributes', 'Job Description'])
```
